```python
column_names = ["Survived", "Pclass", "Embarked", "SibSp"]
```

```python
def showBar(col, group=None):
    if group is None:
        value = df[col]
        value_counts = value.value_counts()
        plt.title(col)
        plt.bar(value_counts.index, value_counts)
        plt.show()
    else:
        value = df[[col, group]]
        value_counts = value.groupby([group]).value_counts().plot.bar()
        plt.show()

#showBar("Survived", "Sex")
for columns in column_names:
    showBar(columns, "Sex")
```
