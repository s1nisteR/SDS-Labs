Get total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties: â€“

Add a circle marker.

```python
# Code Here
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("company_sales_data.csv") #data from csv file
x = data["month_number"].to_numpy()          #convert to numpy array before trying to plot it
y = data["total_profit"].to_numpy()          #convert to numpy array before trying to plot it


plt.plot(x, y, linestyle="dashed", linewidth=3.0, marker="o", color="red", markeredgecolor="red", markerfacecolor="black")
plt.title("Company Sales Data of Last Year")
plt.legend(["Profit data of last year"], frameon=True, loc=4)
plt.xlabel("Month Number")
plt.ylabel("Profit in Dollar")
plt.xticks(range(1, 13))

plt.show()
```

# Exercise 3: 
Read face cream and facewash product sales data and show it using the bar chart
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

The bar chart should look like this.

image.png

```python
# Code Here
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("company_sales_data.csv")

x = data["month_number"].to_numpy()
y = data["facecream"].to_numpy()
y2 = data["facewash"].to_numpy()

plt.bar(x + 0.2, y, 0.4, color="red")
plt.bar(x - 0.2, y2, 0.4, color="darkblue")
plt.xticks(range(1, 13), rotation=90)
plt.legend(["Facecream Sales Data", "Facewash Sales Data"], frameon=True, loc="upper left", fontsize="8")
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.title("Facewash and Facecream sales data")
plt.grid(linestyle="--")
plt.show()
```

# Exercise 4:
Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

The Pie chart should look like this.

image.png

```python
# Code Here
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("company_sales_data.csv")
sales = []
#data.iloc[:, 6]

for i in range(1, data.shape[1] - 2):
    sales.append(data.iloc[:, i].sum())

myLabels = ["Face Cream", "Face Wash", "Toothpaste", "Bathing Soap", "Shampoo", "Moisturizer"]

plt.pie(sales, labels=myLabels, autopct='%1.1f%%')
plt.legend(["Face Cream", "Face Wash", "Toothpaste", "Bathing Soap", "Shampoo", "Moisturizer"], frameon=True, loc=8, fontsize="8")
plt.title("Sales Data")
plt.show()

    


```

# Exercise 5:
Plot Sine and Cosine Graph.

The plot graph should look like this.

image.png

```python
# Code Here
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-1*np.pi, np.pi, 0.1)
y = np.sin(x)

x2 = np.arange(-1*np.pi, np.pi, 0.1)
y2 = np.cos(x2)


plt.plot(x, y, "g-", x2, y2, "b-")
plt.xticks([-1 * np.pi, -1 * np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.gca().set_xticklabels(["-$\pi$", "-$\pi$/2", "0", "$\pi$/2", "+$\pi$"])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Sine and Cosine Graphs")
plt.legend(["sine", "cosine"], frameon=True, loc=0)
plt.margins(y=0)
plt.show()
```
