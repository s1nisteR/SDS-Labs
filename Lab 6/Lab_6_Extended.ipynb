```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Bangladesh_Rain_dataset.csv")
```

# Task 1

```python
filter = df.Year.unique()[-10:]
ten_years = df[df.Year.isin(filter)]
ten_years = ten_years[["Month", "rain"]]
monthwise_mean = ten_years.groupby("Month").mean()
monthwise_mean.plot.bar(title="Monthwise average rainfall for last 10 years")
plt.show()

pie_values = np.array(monthwise_mean["rain"])
mylabels = [i for i in range(1, 13)]
plt.pie(pie_values)
plt.legend(labels=mylabels)
plt.show()
```

As we can see, the month with the highest rainfall were 7 and 8, whereas the month with the lowest rainfall were 1 and 12.

# Task 2

```python
filter = df.Year.unique()[-20:]
twenty_years = df[df.Year.isin(filter)]
twenty_years = twenty_years[["Month", "tem"]]
monthwise_mean = twenty_years.groupby("Month").mean()
months = [i for i in range(1, 13)]
plt.scatter(months, monthwise_mean["tem"])
plt.title("Average Temperature Per Month for Last Twenty Years")
plt.xlabel("Month")
plt.ylabel("Average Temperature")
plt.xticks(months)
plt.grid()
plt.show()
```

The average temperature is highest for month 6 and the lowest for month 1. Usually higher temperatures are more concentrated between months 5 and 9.

# Task 3

```python
filter = df.Year.unique()[-15:]
fifteen_years = df[df.Year.isin(filter)]
fifteen_years.drop(columns=["Year"], inplace=True)
sns.boxplot(x=fifteen_years["rain"])
```

The median rainfall is around 150 to 200, with values ranging from 0(lowest) to around 750 to 780(highest). There are two outliers in the range 850 to 1000. Rainfall data is positively skewed or right skewed.

```python
sns.boxplot(x=fifteen_years["tem"])
```

The median temperature is around 27.5. The minimum is below 18 and maximum is above 28. From the boxplot we cannot observe any outliers. Temprature data is negatively skewed or left skewed.
