# Finding Relationships between Data

```python
df[["age", "income"]].groupby("income").mean().plot.bar(title="Average Age Per Income Group")
```

Finding the age distribution of the data.

We can see that the age distribution is right-skewed, hence there is more people of younger age than older.

```python
df["age"].hist()
```

```python
df[["gender", "income"]].groupby("income").value_counts().plot.bar(title="Gender and Income Relationship")
```

We can see that for both male and female, there is a larger number whose income is less than or equal to 50k rather than more than 50k.

However, from the above we cannnot conclusively determine whether male or female are paid more because: (look below)

```python
df["gender"].value_counts().plot.bar(title="Number of adults in each gender group")
```

we can see that the dataset contains more number of males than females from above. We could however conclusively determine that the dataset contains more adult males than females, which could indicate that more males were hired.

```python
df[["education", "income"]].groupby("income").value_counts().plot.bar(title="Relationship between Education Level and Income")
```

```python

df[["occupation", "income"]].groupby("income").value_counts().plot.bar()
```
